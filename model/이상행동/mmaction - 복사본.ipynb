```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
from mmaction.apis.inferencers import MMAction2Inferencer
import time
from IPython.display import display, clear_output

# 모델 및 비디오 경로
config_path = 'C:/Users/FORYOUCOM/Desktop/school/tsm_imagenet-pretrained-r50_8xb16-1x1x8-100e_kinetics400-rgb.py'
checkpoint_path = 'C:/Users/FORYOUCOM/Desktop/school/tsm_best_acc_epoch7.pth'
video_path = 'C:/Users/FORYOUCOM/Desktop/school/steal.mp4'

# 파라미터 설정
SEQUENCE_LENGTH = 1  # 한 번에 입력할 프레임 수
HYPER_VALUE_STEAL = 0.9  # 특정 행동을 판단하는 임계값
SPEED_MULTIPLIER = 2  # 배속 (1 = 원래 속도, 2 = 2배속)

def get_ratio(model):
    cap = cv2.VideoCapture(video_path)
    frames = deque(maxlen=SEQUENCE_LENGTH)

    if not cap.isOpened():
        print(f"Error: Cannot open video file {video_path}")
        return

    print(f"Processing video at {SPEED_MULTIPLIER}x speed...")

    plt.figure(figsize=(8, 6))

    while True:
        # 배속을 위해 프레임 여러 개 건너뛰기
        for _ in range(SPEED_MULTIPLIER):
            isValidCapture, frame = cap.read()
            if not isValidCapture:
                break  # 영상 끝

        if not isValidCapture:
            break

        frames.append(frame)
        if len(frames) != SEQUENCE_LENGTH:
            continue  # 충분한 프레임이 쌓일 때까지 대기

        # 모델에 입력
        results = model(inputs=np.array(frames))

        # 예측 확률 가져오기
        preds_none, preds_steal = results["predictions"][0]["rec_scores"][0]

        # 프레임에 텍스트 추가
        text = f"Steal Probability: {round(preds_steal, 4)}"
        color = (0, 255, 0) if preds_steal < HYPER_VALUE_STEAL else (0, 0, 255)  # 빨간색 경고
        cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

        if preds_steal > HYPER_VALUE_STEAL:
            cv2.putText(frame, "⚠️ Action Detected!", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # BGR -> RGB 변환 (Matplotlib은 RGB 포맷을 사용)
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Matplotlib을 이용해 프레임 표시
        clear_output(wait=True)
        plt.imshow(frame)
        plt.axis("off")  # 축 없애기
        plt.title(text, fontsize=12)
        display(plt.gcf())

        # 재생 속도 조절 (빠르게 진행)
        time.sleep(0.001)  # 2배속 (기본값 0.03 → 0.015로 줄임)

    cap.release()
    print("Processing complete.")

def main():
    # MMACTION2 모델 로드
    model = MMAction2Inferencer(
        rec=config_path,
        rec_weights=checkpoint_path,
        device="cuda:0",
        input_format="array"
    )

    # 영상 분석 시작
    get_ratio(model)

if __name__ == '__main__':
    main()
```

```python
import cv2
import numpy as np
from collections import deque
from mmaction.apis.inferencers import MMAction2Inferencer
import time

# 모델 및 파라미터 경로
config_path = 'C:/Users/FORYOUCOM/Desktop/school/tsm_imagenet-pretrained-r50_8xb16-1x1x8-100e_kinetics400-rgb.py'
checkpoint_path = 'C:/Users/FORYOUCOM/Desktop/school/tsm_best_acc_epoch7.pth'

# 파라미터 설정
SEQUENCE_LENGTH = 1  
HYPER_VALUE_STEAL = 0.9  
SPEED_MULTIPLIER = 1      # 실시간에서는 1을 사용하는 것이 좋습니다.
DELAY_TIME = 0.05         # 프레임 간 대기 시간

def get_ratio(model):
    # 웹캠 사용 (0번 인덱스; 다른 카메라가 있다면 숫자를 변경)
    cap = cv2.VideoCapture(0)
    frames = deque(maxlen=SEQUENCE_LENGTH)

    if not cap.isOpened():
        print("Error: 웹캠을 열 수 없습니다!")
        return

    print("웹캠 영상을 처리합니다...")

    while True:
        # SPEED_MULTIPLIER 값만큼 프레임을 읽음 (실시간이면 보통 1)
        for _ in range(SPEED_MULTIPLIER):
            ret, frame = cap.read()
            if not ret:
                break

        if not ret:
            break

        frames.append(frame)
        if len(frames) != SEQUENCE_LENGTH:
            continue

        # 모델 추론 수행 (프레임 배열을 numpy array로 전달)
        results = model(inputs=np.array(frames))
        # 예시에서는 두 개의 클래스(예: None과 Steal)가 있다고 가정합니다.
        preds_none, preds_steal = results["predictions"][0]["rec_scores"][0]

        # 추론 결과 텍스트 및 색상 결정
        text = f"Steal Probability: {round(preds_steal, 4)}"
        color = (0, 255, 0) if preds_steal < HYPER_VALUE_STEAL else (0, 0, 255)
        cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

        # 특정 확률 이상일 때 빨간색 경고 오버레이 추가
        if preds_steal > HYPER_VALUE_STEAL:
            cv2.putText(frame, "⚠️ Stealing Detected!", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)
            red_overlay = np.zeros_like(frame, dtype=np.uint8)
            red_overlay[:] = (0, 0, 255)
            frame = cv2.addWeighted(frame, 0.7, red_overlay, 0.3, 0)

        cv2.imshow("Action Recognition", frame)
        time.sleep(DELAY_TIME)

        # q 키를 누르면 종료
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

def main():
    model = MMAction2Inferencer(
        rec=config_path,
        rec_weights=checkpoint_path,
        device="cuda:0",
        input_format="array"
    )
    get_ratio(model)

if __name__ == '__main__':
    main()
```

```python
import os
import cv2
import xml.etree.ElementTree as ET
import glob

def parse_buying_frames(xml_path):
    if not os.path.exists(xml_path):
        print(f"[WARN] XML 파일이 존재하지 않습니다: {xml_path}")
        return None, None
    
    tree = ET.parse(xml_path)
    root = tree.getroot()

    buy_start_frames = []
    buy_end_frames = []

    for track in root.findall('track'):
        label_name = track.get('label', '')
        if label_name == 'buying_start':
            for box_or_points in track.findall('box') + track.findall('points'):
                frame_num = int(box_or_points.get('frame'))
                buy_start_frames.append(frame_num)
        elif label_name == 'buying_end':
            for box_or_points in track.findall('box') + track.findall('points'):
                frame_num = int(box_or_points.get('frame'))
                buy_end_frames.append(frame_num)

    if not buy_start_frames or not buy_end_frames:
        return None, None

    buy_start = min(buy_start_frames)
    buy_end = max(buy_end_frames)
    
    if buy_start > buy_end:
        buy_start, buy_end = buy_end, buy_start

    return buy_start, buy_end

def cut_video_by_frames(video_path, start_frame, end_frame, output_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"[WARN] 영상 열기 실패: {video_path}")
        return False

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0
    success = False

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        if start_frame <= frame_count <= end_frame:
            out.write(frame)
            success = True

        frame_count += 1
        if frame_count > end_frame:
            break

    cap.release()
    out.release()
    if success:
        print(f"[INFO] {output_path} 저장 완료. (구간: {start_frame} ~ {end_frame})")
    else:
        print(f"[INFO] {output_path} 구간이 유효하지 않아 파일이 생성되지 않았습니다.")
    return success

def process_one_video(video_path, xml_path, output_dir):
    """
    비디오+XML → 구매/미구매 구간들을 잘라내어 MP4 생성.
    결과 (clip_path, label) 리스트 반환 (1=구매, 0=미구매).
    """
    clip_label_list = []

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"[WARN] 영상을 열 수 없음: {video_path}")
        return clip_label_list
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()

    # 구매 구간 파싱
    buy_start, buy_end = parse_buying_frames(xml_path)

    base_name = os.path.splitext(os.path.basename(video_path))[0]

    # 구매 구간이 없으면 전체를 미구매(0)로 처리
    if buy_start is None or buy_end is None:
        out_whole_nonbuy = os.path.join(output_dir, f"{base_name}_non_buying_whole.mp4")
        success = cut_video_by_frames(video_path, 0, total_frames - 1, out_whole_nonbuy)
        if success:
            clip_label_list.append((out_whole_nonbuy, 0))
        return clip_label_list

    # 구매 구간(1)
    out_buy = os.path.join(output_dir, f"{base_name}_buying.mp4")
    success_buy = cut_video_by_frames(video_path, buy_start, buy_end, out_buy)
    if success_buy:
        clip_label_list.append((out_buy, 1))

    # 앞쪽 미구매 구간(0)
    if buy_start > 0:
        out_non_buy_front = os.path.join(output_dir, f"{base_name}_non_buying_front.mp4")
        success_front = cut_video_by_frames(video_path, 0, buy_start - 1, out_non_buy_front)
        if success_front:
            clip_label_list.append((out_non_buy_front, 0))

    # 뒤쪽 미구매 구간(0)
    if buy_end < total_frames - 1:
        out_non_buy_back = os.path.join(output_dir, f"{base_name}_non_buying_back.mp4")
        success_back = cut_video_by_frames(video_path, buy_end + 1, total_frames - 1, out_non_buy_back)
        if success_back:
            clip_label_list.append((out_non_buy_back, 0))

    return clip_label_list

def find_video_xml_pairs(root_dir):
    """
    root_dir 아래에 있는 모든 .mp4 파일을 찾고,
    같은 경로에 .xml 확장자가 존재하면 (video_path, xml_path) 형태로 리스트를 만든다.
    """
    pairs = []
    mp4_files = glob.glob(os.path.join(root_dir, "**", "*.mp4"), recursive=True)
    for mp4_path in mp4_files:
        base, _ = os.path.splitext(mp4_path)
        xml_path = base + ".xml"
        if os.path.exists(xml_path):
            pairs.append((mp4_path, xml_path))
        else:
            print(f"[WARN] XML 파일을 찾지 못했습니다: {xml_path}")
    return pairs

def main():
    # 1) 영상+XML이 들어있는 폴더
    root_dir = "E:/test/videos(val)"
    # 2) 잘라낼 결과를 저장할 폴더
    output_dir = "E:/test/val"
    # 3) 최종 TXT 파일 경로
    txt_path = os.path.join(output_dir, "val_label_data.txt")

    # (영상, XML) 쌍 찾기
    video_xml_pairs = find_video_xml_pairs(root_dir)
    print("[INFO] 발견된 (영상, XML) 쌍 수:", len(video_xml_pairs))

    # 결과 (clip_path, label)를 누적할 리스트
    all_clips_labels = []

    # 각 파일 쌍을 처리
    for video_path, xml_path in video_xml_pairs:
        print(f"[INFO] 처리 시작: {video_path}")
        clip_label_list = process_one_video(video_path, xml_path, output_dir)
        all_clips_labels.extend(clip_label_list)

    # TXT로 기록 (파일명, ID) 형태
    os.makedirs(output_dir, exist_ok=True)
    with open(txt_path, 'w', encoding='utf-8') as f:
        # 헤더 대신 첫 줄에 설명 코멘트를 적고 싶다면 (주석형태) 사용 가능
        f.write("# filename label\n")
        
        for clip_path, label in all_clips_labels:
            filename = os.path.basename(clip_path)  # 파일명만 추출
            # 예: "video1_buying.mp4 1" 형태로 기록
            f.write(f"{filename} {label}\n")

    print(f"[DONE] 모든 영상 처리 완료. 라벨 TXT: {txt_path}")

if __name__ == "__main__":
    main()
```

```python
import os
import glob

# 1) txt 파일 경로 및 MP4가 있는 폴더 경로 지정
txt_file_path = "E:/test/make/val/val_label_data.txt"   # 예: "C:/data/train_label_data.txt"
mp4_folder_path = "E:/test/make/val"                      # 예: "C:/data/videos"

# 2) TXT에서 파일명 읽어오기
txt_filenames = set()
with open(txt_file_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        # 공백이거나 빈 줄이면 건너뛰기
        if not line:
            continue
        # 각 줄은 "파일명.mp4 라벨" 식으로 가정
        # 공백으로 split → 첫 번째 토큰이 파일명
        splitted = line.split()
        filename = splitted[0]

        # 확장자가 .mp4인지 확인하여 수집
        if filename.endswith(".mp4"):
            txt_filenames.add(filename)

# 3) 실제 폴더 안에 있는 mp4 파일명 수집
#    glob.glob을 사용하여 폴더 내 *.mp4 파일을 모두 찾음
actual_mp4_paths = glob.glob(os.path.join(mp4_folder_path, "*.mp4"))
# 실제 파일명만 추출
actual_mp4_filenames = set(os.path.basename(path) for path in actual_mp4_paths)

# 4) 비교
#    1) txt에는 있는데 실제 폴더에는 없는 파일
missing_files = txt_filenames - actual_mp4_filenames
#    2) 실제 폴더에는 있는데 txt에는 없는 파일
unlisted_files = actual_mp4_filenames - txt_filenames

# 5) 결과 출력
if not missing_files and not unlisted_files:
    print("TXT와 실제 폴더에 있는 mp4 파일이 완벽히 일치합니다!")
else:
    if missing_files:
        print("[TXT에만 있고, 실제 폴더에는 없는 파일]")
        for f in missing_files:
            print(" -", f)
    if unlisted_files:
        print("[실제 폴더에만 있고, TXT에는 없는 파일]")
        for f in unlisted_files:
            print(" -", f)
```

모델 성능 그래프 만드는 코드드

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report

# 단순한 모델 구조 예시 (checkpoint와 실제 구조가 다를 수 있음)
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.layer = nn.Linear(10, 2)  # 예: 10 입력, 2 출력 (이진 분류)

    def forward(self, x):
        return self.layer(x)

# 모델 로드: checkpoint 파일에서 state_dict 불러오기 (strict=False 옵션 사용)
model = MyModel()
checkpoint = torch.load('C:/Users/FORYOUCOM/Desktop/대학교/졸업작품/school 수정 버전/tsm_imagenet-pretrained-r50_8xb16-1x1x8-50e_kinetics400-rgb_20220831-64d69186.pth', map_location=torch.device('cpu'))
model.load_state_dict(checkpoint['state_dict'], strict=False)
model.eval()

# 테스트 데이터 준비 (예시: 실제 데이터셋으로 대체하세요)
test_data = np.random.randn(100, 10).astype(np.float32)  # 100개 샘플, 10 특성
test_labels = np.random.randint(0, 2, 100)                # 0 또는 1의 실제 레이블

# 데이터를 텐서로 변환
test_tensor = torch.tensor(test_data)

# 모델 예측 수행
with torch.no_grad():
    outputs = model(test_tensor)
    # 로짓이 출력될 경우, softmax를 적용하여 확률값 계산 (이진 분류)
    probabilities = torch.softmax(outputs, dim=1).numpy()
    predictions = np.argmax(probabilities, axis=1)

# 모델 성능 평가: 분류 리포트 출력
print("Classification Report:")
print(classification_report(test_labels, predictions))

# 혼동 행렬(Confusion Matrix) 시각화
cm = confusion_matrix(test_labels, predictions)
plt.figure()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['Class 0', 'Class 1'])
plt.yticks(tick_marks, ['Class 0', 'Class 1'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# ROC 곡선 그리기 (이진 분류: class 1의 확률 사용)
fpr, tpr, _ = roc_curve(test_labels, probabilities[:, 1])
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
